```python
from keras.datasets import fashion_mnist
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
description = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle Boot"]
```

```python
unique_labels = []
images = []
for i in range(len(y_train)):
    if len(unique_labels)==10:
        break;
    if y_train[i] not in unique_labels:
        unique_labels.append(y_train[i])
        images.append(x_train[i])
        
images_temp = [x for _,x in sorted(zip(unique_labels,images))] #sorting images according to values in unique_labels
wandb.log({"Examples": [wandb.Image(img, caption=caption) for img, caption in zip(images_temp, description)]})
plt.figure(figsize=(10,5))
for i in range(1, 12):
    if i==11: 
        break;
    plt.subplot(2, 5, i)
    plt.imshow(images_temp[i-1], cmap=plt.get_cmap('Blues'))
    plt.title(description[i-1])
#Display the entire plot
plt.show()
```
